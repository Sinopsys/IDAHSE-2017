#### Сравнение RF, DT и Bagging'а (2 балла)

На кросс-валидации по обучающей выборке сравните качество и скорость
1. Обычного дерева решений 
2. Bagging'a над деревьями решений с 10, 50, 100, 200 и 300 деревьями
3. Cлучайного леса с 10, 50, 100, 200 и 300 деревьями


* Изобразите полученные результаты на одном графике
* Используйте `StratifiedKFold` с 4 фолдами для оценки качества
* Мера качества - `accuracy`

```python
kf = StratifiedKFold(labels, n_folds=4)
```

```python
def fit_predict_print(clf):
    s = clf[0].__class__.__name__
    
    if (clf[1] > 0):
        s += '; TREE_NUM = ' + str(clf[1])
    print(s)
    accs = []
    for train_index, test_index in kf:
        X_tr, X_ts = train.values[train_index], train.values[test_index]
        y_tr, y_ts = labels[train_index], labels[test_index]
        clf[0].fit(X_tr, y_tr)
        train_predictions = clf[0].predict(X_ts)
        accs.append(accuracy_score(y_ts, train_predictions))
    acc = np.mean(accs)
    print('Accuracy: {:.4%}\n'.format(acc))
    return acc
```

```python
acc_scores = []
classifiers = []
classifiers.append((DecisionTreeClassifier(random_state=RANDOM_SEED), 0))

tree_num = [10, 50, 100, 200, 300]
for n in tree_num:
    bc = BaggingClassifier(n_estimators=n, random_state=RANDOM_SEED)
    rf = RandomForestClassifier(n_estimators=n, random_state=RANDOM_SEED)
    classifiers.append((bc, n))
    classifiers.append((rf, n))
```

```python
for clf in classifiers:
    acc_scores.append(fit_predict_print(clf))
```

```python
backup = acc_scores
```

```python
dt_acc = [acc_scores[0]]*5
bc_acc = []
rf_acc = []
for i in range(10):
    if i % 2 == 0:
        bc_acc.append(acc_scores[1:][i])
    else:
        rf_acc.append(acc_scores[1:][i])
# Translate to percentage
#
dt_acc = [x * 100 for x in dt_acc]
bc_acc = [x * 100 for x in bc_acc]
rf_acc = [x * 100 for x in rf_acc]
```

```python
fig = plt.figure(figsize=(7, 4))
plt.xlim([0, 310])
plt.title('Accuracy of three classifiers')
plt.xlabel('# of trees (not regarding decition tree)')
plt.ylabel('Accuracy, %')
plt.plot(tree_num, dt_acc, label='Decision Tree')
plt.plot(tree_num, bc_acc, label='Bagging Classifier', marker='o')
plt.plot(tree_num, rf_acc, label='Random Forest', marker='o')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Видно, что Random Forest показывает наивысший accuracy

## PCA (1 балл)

* Используйте метод главных компонент, чтобы перейти к пространству меньшей размерности. Подберите количество компонент так, чтобы доля объясненной дисперсии составляла 60%

#### Hint
Составьте [`Pipeline`](http://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html) из PCA и предиктивной модели

```python
variance = 0
X_train_pca = []
n_components = 0
while variance < 0.06:
    X_train_pca = PCA(n_components=n_components, random_state=RANDOM_SEED).fit(X_train)
    variance = X_train_pca.explained_variance_.sum()
    n_components += 1
```

Подобранное кол-во компонент - 13.

```python
X_train_pca.n_components_
```

```python
def get_acc_pl(pl):
    s = pl[1]

    if (pl[2] > 0):
        s += '; TREE_NUM = ' + str(pl[2])
    print(s)
    accs = []
    pl[0].fit(X_train, y_train)
    train_predictions = pl[0].predict(X_test)
    accs.append(accuracy_score(y_test, train_predictions))
    acc = np.mean(accuracy_score(y_test, train_predictions))
    print('Accuracy: {:.4%}\n'.format(acc))
    return acc
```

```python
acc_scores = []
pl_classifiers = []
pl_classifiers.append((Pipeline([('pca', X_train_pca), ('dec', DecisionTreeClassifier(random_state=RANDOM_SEED))]),
                       'DecisionTreeClassifier',
                       0))

for n in tree_num:
    bc = Pipeline([('pca', X_train_pca), ('bag', BaggingClassifier(n_estimators=n, random_state=RANDOM_SEED))])
    rf = Pipeline([('pca', X_train_pca), ('rand', RandomForestClassifier(n_estimators=n, random_state=RANDOM_SEED))])
    pl_classifiers.append((bc, 'BaggingClassifier', n))
    pl_classifiers.append((rf, 'RandomForestClassifier', n))
```

```python
for pl_clf in pl_classifiers:
    acc_scores.append(get_acc_pl(pl_clf))
```

```python
dt_acc_pl = [acc_scores[0]]*5
bc_acc_pl = []
rf_acc_pl = []
for i in range(10):
    if i % 2 == 0:
        bc_acc_pl.append(acc_scores[1:][i])
    else:
        rf_acc_pl.append(acc_scores[1:][i])
# Translate to percentage
#
dt_acc_pl = [x * 100 for x in dt_acc_pl]
bc_acc_pl = [x * 100 for x in bc_acc_pl]
rf_acc_pl = [x * 100 for x in rf_acc_pl]
```

```python
fig = plt.figure(figsize=(7, 4))
plt.xlim([0, 310])
plt.title('Accuracy of three classifiers')
plt.xlabel('# of trees (not regarding decition tree)')
plt.ylabel('Accuracy, %')
plt.plot(tree_num, dt_acc_pl, label='Decision Tree')
plt.plot(tree_num, bc_acc_pl, label='Bagging Classifier', marker='o')
plt.plot(tree_num, rf_acc_pl, label='Random Forest', marker='o')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
