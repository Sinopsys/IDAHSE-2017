```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%pylab inline
```

Функция квантилей

```python
def Q(p):
    return 1/(1-p)
```

Генерация ста равномерно распределенных случайных величин

```python
a = 0
b = 1
uniform_rv = sts.uniform(a, b-a)
y = uniform_rv.rvs(100)
# y = [random.random() for i in range (100)]
```

Подставляем в функцию квантилей - получаем значения функции распределения

```python
cdf = [Q(y_i) for y_i in y]
plt.hist(cdf, bins=10)

plt.ylabel('$F(x)$')
plt.xlabel('$x$');
plt.title('Sample of 100 r.v.')
plt.show()
```

```python
y = uniform_rv.rvs(1000)
```

```python
cdf = [Q(y_i) for y_i in y]
plt.hist(cdf, bins=10)

plt.ylabel('$F(x)$')
plt.xlabel('$x$');
plt.title('Sample of 1000 r.v.')

plt.show()
```

Задание 1, пункт б)

$$Y_i = \sum_{j = 1}^{30} X_{ij} , i = 1..1000$$

```python
Y = []
for i in range (1000):
    Y.append(sum([Q(j) for j in  uniform_rv.rvs(30)]))
```

```python
plt.hist(Y, bins=10)

plt.ylabel('$F(x)$')
plt.xlabel('$x$');
plt.title('Sample of 1000 r.v.')
plt.show()
```

## Задание 2

```python
def task_2(data_len):
    res = {}
    size = 1000
    for e in np.arange(1, 5.1, step=0.5):
        res[e] = 0
        for i in range(size):
            distr = np.random.chisquare(2*e, data_len)
            if sts.ttest_1samp(distr, float(3))[1] <= 0.05:
                res[e] += 1
    x, y= ([], [])
    lst = sorted(res.keys())
    for el in lst:
        x.append(el)
        y.append(float(res[el]) / float(size) * 100)
    plt.plot(x, y, label= 'vol. = ' + str(data_len))
    plt.legend()
```

```python
hyp_0 = 0
hyp_a = 0
for i in range (0, 10000):
    d = np.random.chisquare(6, 8)
    if sts.ttest_1samp(d, 3.0)[1] <= 0.05:
        hyp_a += 1
    else:
        hyp_0 += 1
```

```python
print (hyp_0 // 100, hyp_a // 100)
```

```python
task_2(8)
task_2(50)
plt.show();
```
