<div><p>7. Постройте графики индекса CPI за 2012-2016 для стран своего региона. Примерный вид графика приведен ниже</p>
<img src="img/fig_springfield_region.png">
</div>

```python
def build_plot():
    # dummies для того, чтобы добавить к ним значения 
    #
    score_2012, score_2013, score_2014, score_2015, score_2016 = [], [], [], [], []

    # 5, 10, 15, 20, 25 - это значения на оси X, чтобы точки были не слишком близки
    #
    [score_2012.append([5 , i]) for i in list(df_ssa.loc[:, '2012 Score'].dropna())]
    [score_2013.append([10, i]) for i in list(df_ssa.loc[:, '2013 Score'].dropna())]
    [score_2014.append([15, i]) for i in list(df_ssa.loc[:, '2014 Score'].dropna())]
    [score_2015.append([20, i]) for i in list(df_ssa.loc[:, '2015 Score'].dropna())]
    [score_2016.append([25, i]) for i in list(df_ssa.loc[:, '2016 Score'].dropna())]



    # Задаем рэндж для осей
    #
    plt.xlim([3,27])
    plt.ylim([0,85])

    # Названия графика, осей и т.д.
    #
    plt.title('Sub Saharan Africa CPI for 2012 - 2016')
    plt.ylabel('CPI')
    plt.xticks([5, 10, 15, 20, 25], ['2012 Score', '2013 Score', '2014 Score', '2015 Score', '2016 Score'])
    
    plt.grid(True)
    # Строим точки
    #
    plt.plot(*zip(*score_2012), marker='o', color='m', ls='');
    plt.plot(*zip(*score_2013), marker='o', color='c', ls='');
    plt.plot(*zip(*score_2014), marker='o', color='r', ls='');
    plt.plot(*zip(*score_2015), marker='o', color='g', ls='');
    plt.plot(*zip(*score_2016), marker='o', color='y', ls='');

build_plot()
plt.show()
```

```python
columns = ["Region", "Country", "Score 2012", "Score 2016", "Change in Position"]
df_task = pnd.DataFrame(index=index, columns=columns)
mean_2016 = df_ssa.loc[:, '2016 Score'].mean()
max_2016 = df_ssa.loc[:, '2016 Score'].max()
min_2016 = df_ssa.loc[:, '2016 Score'].min()

# Найдем наиболее близкое к mean_2016 значение
# 
df_ssa_cpy = df_ssa.copy()
df_ssa_cpy['closest'] = df_ssa_cpy['2016 Score'] - mean_2016
closest_value = df_ssa_cpy.loc[:, 'closest'].min()
mean_country = df_ssa_cpy[df_ssa_cpy['closest'] == closest_value].loc[:, 'Country'].values[0]

df_task.loc['mean_2016', 'Country'] = mean_country
df_task.loc['mean_2016', 'Region'] = df_ssa[df_ssa['Country'] == mean_country].loc[:, 'Region'].values[0]
df_task.loc['mean_2016', 'Score 2012'] = df_ssa[df_ssa['Country'] == mean_country].loc[:, '2012 Score'].values[0]
df_task.loc['mean_2016', 'Score 2016'] = df_ssa[df_ssa['Country'] == mean_country].loc[:, '2016 Score'].values[0]
df_task.loc['mean_2016', 'Change in Position'] = df_cpy[df_cpy['Country'] == mean_country].loc[:, 'diff'].values[0]

# Найдем страну обладающую max_2016
# 
max_country = df_ssa[df_ssa['2016 Score'] == max_2016].loc[:, 'Country'].values[0]
df_task.loc['max_2016', 'Country'] = max_country
df_task.loc['max_2016', 'Region'] = df_ssa[df_ssa['Country'] == max_country].loc[:, 'Region'].values[0]
df_task.loc['max_2016', 'Score 2012'] = df_ssa[df_ssa['Country'] == max_country].loc[:, '2012 Score'].values[0]
df_task.loc['max_2016', 'Score 2016'] = df_ssa[df_ssa['Country'] == max_country].loc[:, '2016 Score'].values[0]
df_task.loc['max_2016', 'Change in Position'] = df_cpy[df_cpy['Country'] == max_country].loc[:, 'diff'].values[0]

# Найдем страну обладающую min_2016
# 
min_country = df_ssa[df_ssa['2016 Score'] == min_2016].loc[:, 'Country'].values[0]
df_task.loc['min_2016', 'Country'] = min_country
df_task.loc['min_2016', 'Region'] = df_ssa[df_ssa['Country'] == min_country].loc[:, 'Region'].values[0]
df_task.loc['min_2016', 'Score 2012'] = df_ssa[df_ssa['Country'] == min_country].loc[:, '2012 Score'].values[0]
df_task.loc['min_2016', 'Score 2016'] = df_ssa[df_ssa['Country'] == min_country].loc[:, '2016 Score'].values[0]
df_task.loc['min_2016', 'Change in Position'] = df_cpy[df_cpy['Country'] == min_country].loc[:, 'diff'].values[0]

# Для Russia
# 
ru_country = df[df['Country'] == 'Russia']
df_task.loc['Russia_2016', 'Country'] = ru_country['Country'].values[0]
df_task.loc['Russia_2016', 'Region'] = ru_country['Region'].values[0]
df_task.loc['Russia_2016', 'Score 2012'] = ru_country['2012 Score'].values[0]
df_task.loc['Russia_2016', 'Score 2016'] = ru_country['2016 Score'].values[0]
df_task.loc['Russia_2016', 'Change in Position'] = df_cpy[df_cpy['Country'] == 'Russia'].loc[:, 'diff'].values[0]


df_task
```

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
import os
files = os.listdir('data/stock')
col_names = [s[:-4] for s in files]
df_2 = pd.DataFrame(columns=col_names, index=list(pd.read_csv('data/stock/' + files[0], index_col=0).index))
df_2
for i in range (len(files)):
    df_2[col_names[i]] = pd.read_csv('data/stock/' + files[i], index_col=0, parse_dates=True)['Close']
df_2.drop(['BIDU', 'PYPL'], inplace=True, axis=1)
df_2
```

<p>2. Рассчитайте корреляционную матрицу для всех акций</p>

```python
corrMatrix = df_2.corr()
```

```python
import seaborn as sns
sns.heatmap(corrMatrix, 
            xticklabels=corrMatrix.columns.values,
            yticklabels=corrMatrix.columns.values);
```
