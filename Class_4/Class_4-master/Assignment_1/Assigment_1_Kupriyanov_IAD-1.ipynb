<div><p>7. Постройте графики индекса CPI за 2012-2016 для стран своего региона. Примерный вид графика приведен ниже</p>
<img src="fig_springfield_region.png">
</div>

```python
def build_plot():
    # dummies для того, чтобы добавить к ним значения 
    #
    score_2012, score_2013, score_2014, score_2015, score_2016 = [], [], [], [], []

    # 5, 10, 15, 20, 25 - это значения на оси X, чтобы точки были не слишком близки
    #
    [score_2012.append([5 , i]) for i in list(df_ssa.loc[:, '2012 Score'].dropna())]
    [score_2013.append([10, i]) for i in list(df_ssa.loc[:, '2013 Score'].dropna())]
    [score_2014.append([15, i]) for i in list(df_ssa.loc[:, '2014 Score'].dropna())]
    [score_2015.append([20, i]) for i in list(df_ssa.loc[:, '2015 Score'].dropna())]
    [score_2016.append([25, i]) for i in list(df_ssa.loc[:, '2016 Score'].dropna())]



    # Задаем рэндж для осей
    #
    plt.xlim([3,27])
    plt.ylim([0,85])

    # Названия графика, осей и т.д.
    #
    plt.title('Sub Saharan Africa CPI for 2012 - 2016')
    plt.ylabel('CPI')
    plt.xticks([5, 10, 15, 20, 25], ['2012 Score', '2013 Score', '2014 Score', '2015 Score', '2016 Score'])
    
    plt.grid(True)
    # Строим точки
    #
    plt.plot(*zip(*score_2012), marker='o', color='m', ls='');
    plt.plot(*zip(*score_2013), marker='o', color='c', ls='');
    plt.plot(*zip(*score_2014), marker='o', color='r', ls='');
    plt.plot(*zip(*score_2015), marker='o', color='g', ls='');
    plt.plot(*zip(*score_2016), marker='o', color='y', ls='');

build_plot()
plt.show()
```

```python
# Найдем значения для России
df.loc[df['Country'] == 'Russia', ['2012 Score', '2013 Score', '2014 Score', '2015 Score', '2016 Score']]
```

```python
# Точки строятся точь-в-точь как в предыдущем номере, линии строю отдельно

# Получаю значения из рузультатов предыдущих исследований
#
max_scores = [65, 64, 63, 63, 60]
min_scores = [8, 8, 8, 8, 10]
mean_scores = [42.469880, 41.957831, 42.530120, 42.662651, 42.469880]
rus_scores = [28, 28, 27, 29, 29]

# Значения оси X
#
xaxes = [5, 10, 15, 20, 25]

# Собираем значения осей X и Y
#
max_line, min_line, mean_line, rus_line = [], [], [], []
[max_line.append([xaxes[i], max_scores[j]]) for i in range(len(xaxes)) for j in range(len(max_scores)) if i == j]
[min_line.append([xaxes[i], min_scores[j]]) for i in range(len(xaxes)) for j in range(len(min_scores)) if i == j]
[mean_line.append([xaxes[i], mean_scores[j]]) for i in range(len(xaxes)) for j in range(len(mean_scores)) if i == j]
[rus_line.append([xaxes[i], rus_scores[j]]) for i in range(len(xaxes)) for j in range(len(rus_scores)) if i == j]

# Задаем пределы для графика
#
plt.xlim([3,27])
plt.ylim([0,85])

# Названия
#
plt.title('Sub Saharan Africa CPI for 2012 - 2016')
plt.ylabel('CPI')

# Метод, вызывающий построение графика из прошлого задания
build_plot()

# Чертим линии
#
plt.plot(*zip(*max_line), marker='o', ls='-', label='max');
plt.plot(*zip(*min_line), marker='o', ls='-', label='min');
plt.plot(*zip(*mean_line), marker='o', ls='-', label='mean');
plt.plot(*zip(*rus_line), marker='o', ls='-', label='Rus');

# Уменьшим легенду, чтобы она не закрывала часть графика
#
plt.legend(loc=1, prop={'size':8})

plt.show()
```
